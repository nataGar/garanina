# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('power.csv')
data[:3]
```

```python
filtered_countries = data [ (data['country'] == 'Latvia') | (data['country'] == 'Lithuania') | (data['country'] == 'Estonia')]

filtered_category = filtered_countries[ (filtered_countries['category'] == 4) | (filtered_countries['category'] == 12)|(filtered_countries['category'] == 21)] 

filtered_quantity = filtered_category [filtered_category['quantity'] > 0]

filtered_year = filtered_quantity[(filtered_quantity['year'] >= 2005) & (filtered_quantity['year'] <= 2010)]

#quantity_sum_country = filtered_quantity.groupby('country')['quantity'].sum() # сумма по каждой стране в отдельности

quantity_sum_all = filtered_year['quantity'].sum()                        # сумма по всем странам Прибалтики

#guantity_sum_country  

quantity_sum_all
```

```python
rating = pd.read_csv('ratings.csv')   
movies = pd.read_csv('movies.csv')
rating_movies = pd.merge(rating, movies, on='movieId')  # объединим по столбцу 'movieId'
rating_movies
```

```python
# находим фильм с максимальным количеством оценки 5.0

rating_movies.groupby('title')['rating'].value_counts().loc[lambda x: x == max(x) ].to_frame() 

# почему два раза rating выводит??
```
